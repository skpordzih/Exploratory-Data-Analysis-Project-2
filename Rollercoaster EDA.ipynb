# Analysing The Data

## Question 1
What are the locations with the fastest roller coasters?

```python
df.query('Location != "Other"').groupby('Location')['Speed_mph'].agg(['mean','count']).query('count >=10') \
   .sort_values('mean')
```

```python
#These locations have the most fastest rollercoasters
#Now we will view as a plot 
ax=df.query('Location != "Other"').groupby('Location')['Speed_mph'].agg(['mean','count']).query('count >=10') \
   .sort_values('mean')['mean'].plot(kind='bar',title= 'Average Rollercoaster Speed By Location', figsize=(10,5))

ax.set_xlabel('Location')
ax.set_ylabel('Average Speed')
plt.show()
```

## Question 2
What are the top coaster manufacturers by the number of coasters produced?

```python
manufacturer_counts = df['Manufacturer'].value_counts()
top_manufacturers = manufacturer_counts.head(10)  # Assuming you want the top 10 manufacturers
print("Top Coaster Manufacturers by Number of Coasters Produced:")
top_manufacturers
```

## Question 3
How does the distribution of coaster speeds vary between wooden and steel coasters?

```python
#Here I will compare the distributions of 'Speed_mph' for wooden and steel coasters using histograms and KDE plots.
# First up is to filter the dataframe for wooden and steel coasters
wooden_coasters = df[df['Type_Main'] == 'Wood']
steel_coasters = df[df['Type_Main'] == 'Steel']

# Plotting the histograms for speed distribution of wooden and steel coasters
plt.figure(figsize=(10, 6))
sns.histplot(wooden_coasters['Speed_mph'], bins=20, kde=True, color='brown', label='Wooden Coasters')
sns.histplot(steel_coasters['Speed_mph'], bins=20, kde=True, color='blue', label='Steel Coasters')
plt.title('Distribution of Coaster Speeds')
plt.xlabel('Speed (mph)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

## Question 4
Which type of coaster (wooden, steel, or other) has the highest average height and speed?

```python
# Calculate the average height and speed for each type of coaster
average_height_speed = df.groupby('Type_Main').agg({'Height_ft': 'mean', 'Speed_mph': 'mean'})
average_height_speed
```

```python
# Plotting average height and speed for each type of coaster
ax = average_height_speed.plot(kind='bar', title='Average Height and Speed for Each Coaster Type')
ax.set_xlabel('Coaster Type')
ax.set_ylabel('Average Value')
plt.xticks(rotation=45)
plt.show()
```

## Question 5
Are there any geographical patterns in coaster distribution, such as clustering of coasters in certain regions?

```python
# To analyze the geographical distribution of coasters and look for any clustering patterns, 
# I will create a scatter plot using latitude and longitude coordinates. This will help visualize 
# where the roller coasters are located.

# First up is to filter out rows with available latitude and longitude values
df_geo = df.dropna(subset=['latitude', 'longitude'])

# Plotting the geographical distribution of coasters
plt.figure(figsize=(10, 6))
plt.scatter(df_geo['longitude'], df_geo['latitude'], alpha=0.5)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Geographical Distribution of Roller Coasters')
plt.grid(True)
plt.show()
```

## Question 6
Is there any relationship between the year a coaster was introduced and its speed or height?

Using data from my analysis/plots above:
1. **Scatterplot with Year Introduced**:

4. **Correlation Analysis**:

**Conclusion:**
Based on the scatterplots, pairplot, and correlation analysis, there doesn't seem to be a significant relationship between the year a coaster was introduced and its speed or height. The weak correlations suggest that other factors may have more influence on these characteristics.

## Question 7 
How do coaster speed and height correlate with the number of inversions or g-force?

The scatterplot shows the relationship between coaster speed and height. It seems there might be a positive correlation, suggesting that taller coasters tend to have higher speeds. However, the correlation isn't perfectly clear from the scatterplot alone.

Using data from the Correlation Matrix:
  - **Speed vs. Height**: There is a relatively strong positive correlation of 0.73, indicating that as the speed increases, the height tends to increase as well.
  - **Speed vs. Inversions**: There is a very weak negative correlation of -0.03, suggesting almost no linear relationship between speed and the number of inversions.
  - **Speed vs. G-force**: There is a moderately positive correlation of 0.61, indicating that higher speeds tend to be associated with higher g-forces.
  - **Height vs. Inversions**: There is a very weak negative correlation of -0.08, suggesting almost no linear relationship between height and the number of inversions.
  - **Height vs. G-force**: There is a weak positive correlation of 0.47, indicating that some taller coasters tend to have higher g-forces.

**Conclusion**: Coaster speed tends to correlate more strongly with height and g-force than with the number of inversions.There seems to be a positive relationship between speed, height and g-force, while the relationship with inversions is less clear.
